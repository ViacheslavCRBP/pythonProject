```python
# Ищем в коде ссылку на карту поиска:
# https://www.delivery-club.ru/api/v1/cart?longitude=40.261101&latitude=43.670097&shippingType=delivery&autoTranslate=false

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 YaBrowser/23.1.0.2947 Yowser/2.5 Safari/537.36'}
params = {'latitude': '43.670097',
          'longitude': '40.261101',
          'query': 'чизкейк'}
url = 'https://www.delivery-club.ru/api/v1/cart'

response = requests.get(url, headers=headers, params=params).json()
```

```python
pprint(response)
```
